##變數定義
1.   vote：投票票數，單位(萬)
2.   debut：參賽者最終是否進出道組；是=1，否=0
3.   center：表演時是否為中心；是=1，否=0
4.   search：參賽者在Naver的搜尋趨勢
5.   age：參賽者的年齡
6.   beauty：參賽者們互相評價顏值的排序是否在前11名；是=1，否=0
7.   skill1：參賽者第一次技巧評級；級別為ABCDF，分別得分為54321
8.   skill2：參賽者第二次技巧評級；級別為ABCDF，分別得分為54321
9.   first_rank：節目開播前，參賽者投票排名是否有進出道組；是=1，否=0
10.   reputation：節目開播前，參賽者是否出道過或參加過其他選秀節目；是=1，否=0
11.   korean：參賽者國籍是否為韓國人；是=1，否=0

###分出最終出道者

```
win = pd_form["debut"]==1
idol = pd_form[win]
idol.head()
```

```
idol.shape
```

####票數VS搜尋量

```
grouped =idol.groupby("vote")

naver= grouped["search"].sum().sort_values(ascending = False)#sort_values可以將數值按照大小作排序
naver.index #index 看出搜尋量的數值
```

```
naver.values
```

```
#匯入繪圖模組
import matplotlib.pyplot as plt

figure, axis = plt.subplots(figsize=(20, 6))#調整圖表大小
naver_ht=len(naver.values)
plt.bar(range(naver_ht), naver.values, color="#fdaadc")
plt.xticks(range(naver_ht), naver.index)
plt.title("media's influence")
plt.xlabel("vote")
plt.ylabel("search")
plt.show()
```

當搜尋量越多時，票數相對較高

####票數VS年紀

```
separate_age = idol.groupby("age")
youngold= separate_age["vote"].count().sort_values(ascending= False)
```

```
youngold.index
```

```
youngold. values
```

```
import matplotlib.pyplot as plt

Youngold_ht=len(youngold.values)
plt.bar(range(Youngold_ht), youngold.values, color="#ffc300", alpha=0.7)#alpha 可讓顏色刷淡
plt.xticks(range(Youngold_ht), youngold.index)
plt.title("young or old")
plt.xlabel("age")
plt.ylabel("number of debut")
plt.show()
```

當年紀偏年輕，出道機率就較高

###票數、搜尋量VS顏值

```
hex_colors = ["#40dbdb", "#efc0b8"]
continents = idol["beauty"].unique()
for col, cont in zip(hex_colors, continents):
  subset_df = idol[idol["beauty"] == cont]
  plt.scatter(subset_df["search"], subset_df["vote"], color=col, alpha=1, label=cont)

plt.legend()
plt.show()
```

當顏值較高，票數及搜尋量都較高
